```python
#cleaning data
#year has many npn year values 
car = car[car['year'].str.isnumeric()]   # keeps only  those rows in year column which contains numbers
```

```python
 #year is in object change to it in integer 
 car['year']=car['year'].astype(int)
```

```python
 #price has asked for price 
car=car[car['Price']!='Ask For Price']
```

```python
#price has coomas in its prices and it is in object to change it to an integer
car['year']=car['year'].astype(int)
car['Price']=car['Price'].str.replace(',','').astype(int)
```

```python
#kms driven object values with kms atlast
car['kms_driven']=car['kms_driven'].str.split().str.get(0).str.replace(',','')
```

```python
#it has NaN and two rows have petrol in them
car=car[car['kms_driven'].str.isnumeric()]
```

```python
  car=car.reset_index(drop=True)
```

```python
car=car[car['Price']<6000000]
```

Extracting training data


```python
X=car[['name','company','year','kms_driven','fuel_type']]
y=car['Price']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
# Checking R2 Score
r2_score(y_test,y_pred)     # Finding the model woth a random state of train test split where the model was found to give almost
                            # 0.92 as R2 score
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
pipe.predict(pd.DataFrame(columns=X_test.columns,data=np.array(['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']).reshape(1,5)))
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
